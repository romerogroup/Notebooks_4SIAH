# JupyterLab: Interactive Scientific Computing

The [JupyterLab Interface](https://jupyterlab.readthedocs.io/en/stable/user/interface.html) is an interface for interactive scientific computing. We say it is an interface because JupyterLab will be a middle man in our interaction with the computer. The term Interactive Scientific Computing means that different from a integrated development environment (IDE) that is used to develop code in a programming language, JupyterLab include features that are more suited for exploratory or interactive programming the code is dynamically evolving with the output it produces during its development. 

JupyterLab provides access to iPython notebooks, the place where code is written and executed.
It also provides a **File Browser** where you can see the folder structure of the environment.
Other tools include a **Terminal Window** where Unix/Linux commands can be typed and a basic **text editor**. 

On the top of the interface you will see the **menu bar**, On the **left sidebar**, several icons allow you change betwen a **File Browser**, a list of **Terminals and Kernels**, a presentation of **Table of Contents** of the current notebook and an **Extension Manager**. The major area is the **Main Work Area** that is where you open Jupyter Notebooks, each one as a tab. When first open, JupyterLab shows the "Launcher" page. 

<img src="https://github.com/happidence1/AILabs/blob/master/images/jupyterlab.png?raw=1">

The file browser can be navigated just like any other file explorer. A double click on any of the items will open a new tab with its content.

The main work area includes tabbed views of open files that can be closed, moved, and edited as needed. 

The notebooks, including this one, consist of a series of content and code **cells**.  To execute code in a code cell, press `Shift+Enter` or the "Run" button in the menu bar above, while a cell is highlighted. 

The cell can enter in editing mode by pressing `Enter`. Pressing `Shift+Enter` will switch it back to a readable form.

On the top right corner of each Notebook you will see the name of the Kernel running. A kernel is the interpreter of a particular Programming Language. For the purpose of this workshop, the only language we will be using is Python by keep in mind that JupyterLab is language agnostic, and it could run kernels in other programming languages such as Julia and R.

# Checking the libraries used in this workshop

On this workshop we will be working entirely on the Python Programming Language. The Python Language includes a extensive [Standard Library](https://docs.python.org/3/library/).

In computer programming a Standard Library is a library made available across implementations of a programming language. From the users' perspective, it seems that the standard library is part of the language as it can be used without any worries about the presence of the library during execution. Pythonâ€™s standard library is very extensive, offering a wide range of facilities from low level access to low level IO to high level functions to work with XML and JSON files.

On this workshop we will a few of those Standard Libraries but also other libraries that are separate libraries and are developed separately from the language itself. Seven libraries have been chosen for the purposes of our workshop:

  * **NumPy**: NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.
  
  * **SciPy**: SciPy is a Python library used for scientific computing and technical computing. SciPy contains modules for optimization, linear algebra, integration, interpolation, special functions, FFT, signal and image processing, ODE solvers and other tasks common in science and engineering.
    
  * **Matplotlib**: Matplotlib is de-facto standard plotting library for the Python programming language and its numerical mathematics extension NumPy.
  
  * **Pandas**: Pandas is a Python library for data analysis. Pandas is mainly used analysis and manipulation of tabular data in DataFrames. Pandas allows importing data from various file formats such as comma-separated values (CSV), JSON, Parquet, SQL database tables or queries, and Microsoft Excel.
  
  * **Seaborn**: Seaborn is a Python data visualization library built on top of matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
  
  * **scikit-learn**: scikit-learn os a Python library with efficient tools for predictive data analysis and Machine Learning. scikit-learn includes implementation of classical algorithms for regression, classification clustering and dimensionality reduction.
  
  * **NetworkX**: NetworkX is a Python package for the creation, manipulation, and study of the structure, dynamics, and functions of complex networks. The study of networks and graphs can provide light on relations happening in physical, biological and social entities.
  
From this list we have set aside more specialized libraries such as those for Deep Learning (TensorFlow, PyTorch and MXNet). Libraries for BigData analysis (PySpark and Dask) among many other libraries that could be useful in scientific computing in general and data science in particular.

## SciPy

````python
import numpy as np
from scipy.stats import gamma
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)

a = 1.99
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')

x = np.linspace(gamma.ppf(0.01, a), gamma.ppf(0.99, a), 100)
ax.plot(x, gamma.pdf(x, a), 'r-', lw=5, alpha=0.6, label='gamma pdf');
ax.legend();
````

## Matplotlib

````python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

b = 10;
f = lambda x,y: (x-1)**2 + b*(y-x**2)**2;

X = np.arange(-2, 2, 0.03)
Y = np.arange(-1, 3, 0.03)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

fig = plt.figure(figsize=plt.figaspect(.33))
fig.suptitle('The Rosenbrock function')

################
# First subplot
################
ax = fig.add_subplot(1, 2, 1)
plt.contour(X,Y,Z,100)

#################
# Second subplot
#################
ax = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                       linewidth=0, antialiased=False)

ax.set_zlim3d(-1, 200)

plt.show()
````
